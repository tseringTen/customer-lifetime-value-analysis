```python
df.columns=df.columns.str.lower().str.replace(' ','_')  ## for neat and consistent presentation " SNAKE CASING"
df=df.rename(columns={'customerid':'customer_id'})
df.columns
```

```python
df['total_charges']=pd.to_numeric(df['total_charges'],errors='coerce') ## "Type Casting for aggregation purpose from string to float"
df['total_charges']=df['total_charges'].fillna(0)
```

```python
bins=[0,12,36,df['tenure_months'].max()] ## feature engineering getting customer type from tenure
labels=['New','Established','Loyal']

df['customer_type']=pd.cut(df['tenure_months'],bins=bins,labels=labels,include_lowest=True)
```

```python
labels2=['Low','Medium','High']                               ##feature engineering segregating customers based on ther lifte time value
df['cltv_type']=pd.qcut(df['cltv'],q=3,labels=labels2)
```

```python
df['churn_label_value']=df['churn_label'].map({'Yes': 1,'No':0})
(df['churn_label_value']==df['churn_value']).all()                   ## so churn_label and churn_value is the same info in diff formant ,to avoid redundancy i would drop one.
```

```python
df=df.drop(columns=['churn_label']) # so churn_label column dropped
df=df.drop(columns=['churn_label_value']) #temporaray column dropped
```
